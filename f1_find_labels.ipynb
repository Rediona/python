### process for matching

```python
#create a list variable_names to grab the column names of my dataset

var_names_or = list(data.columns.values)
```

```python
#οριζω df τη λιστα με τα ονοματα των στηλων απτο dataset 
names_or = pd.DataFrame(var_names_or)
# ονομαζω τη στηλη για να μπορω να συγκρινω μετα τα dataset
names_or.columns = ['my_labels']
names_or
```

```python
names_or['my_labels'].str.strip() #removes spaces at begining and end
names_or.head()
```

φτιαχνω δευτερη στηλη οπου αποθηκευω διπλα τις αλλαγες που κανω στην αρχικη.
Στο heliad δεν υπαρχουν '_', μονο σε πολυ λιγα labels που θα τα αλλαξω με το χερι. Εδω αφαιρω τις '_' και τα '', και κραταω τη 1η
λεξη για να συγκρινω με τη heliad


```python
names_or['changed'] = names_or['my_labels'].apply(lambda x: x.split('_')[0])
```

```python
#names_or.drop(labels='Name',axis=1,inplace=True)
names_or.head()
```

```python
#κραταω απο τη heliad μονο τις στηλες με την εξηγηση και το label
heliad_names.drop(['Type','Study','Population','DCE','Dataset'],axis=1,inplace=True)
```

```python
heliad_names['Name'].str.strip() #αφαιρω κενα αν υπαρχουν
heliad_names.tail()
```

```python
#αντικαθιστω με τιποτα τις τελειες γτ στο δικο μου dataset ειναι ενωμενες λεξεις, για να μπορω να συγκρινω. φτιαχνω νεα
#στηλη για τις αλλαγες
heliad_names['changed'] = heliad_names['Name'].apply(lambda x: x.replace('.',''))
heliad_names.head(15)
```

```python
#have kept all rows from my dataset(right)and null where there where no matches
#try to keep the index of my dataset so i can just copy in the excel and have the explanations in the wright order
#match = pd.merge(heliad_names,names_or,on=['changed'],how='right',right_index=True)

match = pd.merge(names_or,heliad_names,on=['changed'],how='left')
```

```python
#αρα εχω τα κοινα labels και διπλα την περιγραφη τους
match
```

```python
match.isnull().values.sum() #ουσιαστικα μονο σε 47 δεν εγινε match, αλλα ειναι 94 γτ εχω την επιπλεον στηλη label στο heliad!!
```
